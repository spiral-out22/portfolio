```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Plotting (aggregated metrics only)


k_value=10
plotting_spearman = True
plotting_ndcgk = True
plotting_map = True

graph_title ='giacomelli dataset'
boltz_filename = 'giacomelli_mean_metrics_boltz2_aggregate.csv'
esm2_filename = 'giacomelli_mean_metrics_esm2_aggregate.csv'

#load data source
mean_df_boltz = pd.read_csv(boltz_filename)
mean_df_esm = pd.read_csv(esm2_filename)
```

```python
#Plot confrontation graphs
color=sns.color_palette("Set2")

if plotting_spearman is True:
    legend_label='Spearman'
    plt.figure(figsize=(12,6))
    boltz_spearman_mean = mean_df_boltz['spearman_mean']
    boltz_spearman_std = mean_df_boltz['spearman_std']
    esm_spearman_mean = mean_df_esm['spearman_mean']
    esm_spearman_std = mean_df_esm['spearman_std']
if plotting_ndcgk is True:
    legend_label='nDCG@k'
    plt.figure(figsize=(12,6))
    boltz_spearman_mean = mean_df_boltz['nDCG_mean']
    boltz_spearman_std = mean_df_boltz['nDCG_std']
    esm_spearman_mean = mean_df_esm['nDCG_mean']
    esm_spearman_std = mean_df_esm['nDCG_std']
if plotting_map is True:
    legend_label='mAP'
    plt.figure(figsize=(12,6))
    boltz_AP_mean = mean_df_boltz['mAP']
    boltz_AP_std = mean_df_boltz['AP_std']
    esm_AP_mean = mean_df_esm['mAP']
    esm_AP_std = mean_df_esm['AP_std']
    


#Boltz data
plt.plot(range(1,11) , boltz_spearman_mean , label=f'{legend_label} boltz2', color=color[3] , marker='^')
plt.fill_between(range(1,11), boltz_spearman_mean - boltz_spearman_std , boltz_spearman_mean + boltz_spearman_std , color=color[3], alpha=0.2)


#ESM data
plt.plot(range(1,11) , esm_spearman_mean , label=f'{legend_label} esm2', color=color[1] , marker='^')
plt.fill_between(range(1,11), esm_spearman_mean - esm_spearman_std , esm_spearman_mean + esm_spearman_std , color=color[1], alpha=0.2)



plt.xticks(range(1,11))
plt.xlabel('Round')
plt.legend()

if plotting_spearman is True:
    plt.title(f'{graph_title} Spearman' , fontsize=10)
    plt.ylim(-1,1)
    plt.xlim(0.8 , 10.2)
    plt.yticks([-1,-0.8,-0.6,-0.4,-0.2,0,0.2,0.4,0.6,0.8,1])
    plt.savefig(graph_title + '_spearmanrho.png' , dpi=400)
    

if plotting_ndcgk is True:
    plt.title(f'{graph_title} nDCG@K={k_value}' , fontsize=10)
    plt.ylim(0.5,1)
    plt.xlim(0.8 , 10.2)
    plt.yticks([0.4,0.6,0.8,1])
    plt.savefig(graph_title + '_nDCG@K.png' , dpi=400)
    

if plotting_map is True:
    plt.title(f'{graph_title} mAP' , fontsize=10)
    plt.ylim(0,1)
    plt.xlim(0.8 , 10.2)
    plt.yticks([0,0.2,0.4,0.6,0.8,1])
    plt.savefig(graph_title + '_mAP.png' , dpi=400)
    
```
